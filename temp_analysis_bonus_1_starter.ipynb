```python
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
```

```python
# "tobs" is "temperature observations"
df = pd.read_csv('../Resources/hawaii_measurements.csv')
df.head()
```

```python
# Convert the date column format from string to datetime
df['date']=pd.to_datetime(df['date'])
```

```python
# Set the date column as the DataFrame index
df1 = df.set_index('date')
```

```python
# Drop the date column
datedropped_df = df.drop('date', axis=1)
```

```python
### Compare June and December data across all years 
june_df = df[df['date'].dt.month==6]
dec_df = df[df['date'].dt.month==12]

june_df_plot = june_df.drop(['station','date'], axis=1)
dec_df_plot = dec_df.drop(['station','date'], axis=1)

fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(8,6))
june_df_plot.plot(ax = axes[:,0],subplots=True)
dec_df_plot.plot(ax = axes[:,1], subplots=True)

```

```python
from scipy import stats
```

```python
# Filter data for desired months
```

```python
# Identify the average temperature for June
avg_temp_june = june_df['tobs'].mean()
avg_temp_june
```

```python
# Identify the average temperature for December
avg_temp_dec = dec_df['tobs'].mean()
avg_temp_dec
```

```python
# Create collections of temperature data
temp_col_dec_df = pd.DataFrame(dec_df.iloc[0:1517,3])
temp_col_june_df = pd.DataFrame(june_df.iloc[0:1517,3])
```

```python
# Run paired t-test
stats.ttest_rel(temp_col_dec_df['tobs'], temp_col_june_df['tobs'])
```

### Analysis

The p-value is extremely small (far below 0.05) which indicates it has statistical significance and gives strong evidence against null hypothesis. The results are not random.
